```
class Entities:
  
  def __init__(self, result, text):
    self. result = result
    self.text = text
    self.ents = dict()
    
    self.obtain_ents()

  def obtain_ents(self):
    self.ents = [{"start": ent['start'],
                              "end": ent['end'],
                              "label": ent['entity_group']
                  } for ent in self.result if ent['entity_group'] != '0']
    
  def parse_ner_spacy(self):
    dict_results = {
        "text": self.text,
        "ents": self.ents,
        "title": None
    }
    return dict_results
  
  
```

# Model preparation

```
MAX_LEN = 128
PRETRAINED_MODEL = 'microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract-fulltext'
```

```
class BioBert_NER(nn.Module):

   def __init__(self, n_classes):
       super(BioBert_NER, self).__init__()
       self.config = AutoConfig.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract-fulltext", finetuning_task="ner",base_model_prefix='ner')
       self.bert = BertModel.from_pretrained(PRETRAINED_MODEL, config=config)
       self.out = nn.Linear(self.bert.config.hidden_size, n_classes)

   def forward(self, input_ids, attention_mask):
      #  _, output = self.bert(**input)
      _, pooled_output = self.bert(
      input_ids=input_ids,
      attention_mask=attention_mask, return_dict=False)
      out = self.out(pooled_output)
      return out
```

```
tokenizer = AutoTokenizer.from_pretrained(PRETRAINED_MODEL)
```

```
NER_model = BioBert_NER(4)
NER_model = NER_model.to(device)
```
